```python
from yt_dlp import YoutubeDL
from pprint import pprint

dlopts = {
    'format': 'worstvideo',
    'writesubtitles': 'true',
    'subtitleslangs': ['en'],
    'subtitlesformat': 'vtt',
    'writeautomaticsub': 'true',
}

with YoutubeDL(dlopts) as ydl:
    ydl.download(["https://www.youtube.com/watch?v=R1NSU5TqxU0", "https://www.youtube.com/watch?v=cj--c6VAryg"])
```

```python
import re
from pprint import pprint
from typing import List

def trim_normal_subs(subtitles_filename: str) -> str:
    with open(subtitles_filename, "r") as subs:
        results = "\n".join(list(filter(lambda y: y != '\n', map(lambda x: re.sub(r'^\d{2}:\d{2}:\d{2}\.\d{3}.*', '', x),subs.readlines()[4:]))))
    return results
```

```python
def trim_auto_subs(subtitles_filename: str) -> str:
    with open(subtitles_filename, "r") as subs:
        results = filter(lambda y: y != '\n', map(lambda x: re.sub(r'^\d{2}:\d{2}:\d{2}\.\d{3}.*', '', x),subs.readlines()[4:]))
        results = "\n".join(list(map(lambda x: re.sub(r'<\d{2}:\d{2}:\d{2}.\d{3}>', '', x), results)))
    return results
```

```python
print(trim_auto_subs('How To Make Persian Beef Koobideh Kebab [R1NSU5TqxU0].en.vtt'))
```

```python
print(trim_normal_subs("YOUTUBE DRAMA _PART 2 [cj--c6VAryg].en.vtt"))
```
